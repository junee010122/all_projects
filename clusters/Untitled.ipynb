```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# NMOS Only Configuration
ax1.plot([1, 5], [1, 1], 'k-')  # Long connection
ax1.add_patch(patches.Circle((1, 1), 0.1, color='blue', label='Driver'))
ax1.add_patch(patches.Circle((5, 1), 0.1, color='red', label='GaN HEMT'))
ax1.text(2.5, 1.2, 'Long Electrical Path\nHigh Parasitic Inductance', ha='center')
ax1.set_xlim(0, 6)
ax1.set_ylim(0, 2)
ax1.set_title('NMOS Only')
ax1.axis('off')

# CMOS Integration Configuration
ax2.plot([1, 2], [1, 1], 'k-')  # Short connection
ax2.add_patch(patches.Circle((1, 1), 0.1, color='blue', label='Driver'))
ax2.add_patch(patches.Circle((2, 1), 0.1, color='red', label='GaN HEMT'))
ax2.text(1.5, 1.2, 'Short Electrical Path\nLow Parasitic Inductance', ha='center')
ax2.set_xlim(0, 3)
ax2.set_ylim(0, 2)
ax2.set_title('CMOS Integration')
ax2.axis('off')

plt.legend(handles=[patches.Patch(color='blue', label='Driver'), patches.Patch(color='red', label='GaN HEMT')])
plt.show()
```
