```python
def plot_data(x_vals, y_vals, title, 
              x_label, y_label, y_lim=None, 
              fig_size=(10, 5), font_size=14):

    fig, ax = plt.subplots(figsize=fig_size)

    ax.plot(x_vals, y_vals, linewidth=5)

    ax.set_title("%s" % title)
    ax.set_xlabel("%s" % x_label)
    ax.set_ylabel("%s" % y_label)

    if y_lim is not None:
        ax.set_ylim(y_lim)

    fig.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def load_data(filepath):
    # Replace this function with your actual data loading logic
    return pd.read_csv(filepath)

# Assuming 'load_data' and 'path_list' are properly defined, update with your actual file paths
target_names = ['train_loss_epoch', 'valid_MAE_epoch', 'valid_MSE_epoch', 'valid_emd_epoch']

# Initialize data structures
all_data1 = {name: [] for name in target_names}
all_data2 = {name: [] for name in target_names}

# Load data from the first file in the list for dataset1 and dataset2
data1 = load_data(path_list[0])
data2 = load_data(path_list[0])

# Process data1
for name in target_names:
    if name in data1.columns:
        df = data1.dropna(subset=[name])
        tag = name.split("_")[-1]
        x_vals = df[tag] if tag in df.columns else list(range(df.shape[0]))
        y_vals = df[name]
        all_data1[name].append((x_vals, y_vals, 'Dataset 1'))

# Process data2
for name in target_names:
    if name in data2.columns:
        df = data2.dropna(subset=[name])
        tag = name.split("_")[-1]
        x_vals = df[tag] if tag in df.columns else list(range(df.shape[0]))
        y_vals = df[name]
        all_data2[name].append((x_vals, y_vals, 'Dataset 2'))

# Create a 1x4 subplot array
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
fig.suptitle('Comparative Plots of Dataset 1 and Dataset 2')

# Plotting the data
for ax, name in zip(axes, target_names):
    if all_data1[name] or all_data2[name]:
        ax.set_title(name)
        ax.set_xlabel("Epoch" if "lr" in name else name.split("_")[-1])
        ax.set_ylabel(name)
        for x_vals, y_vals, label in all_data1[name]:
            ax.plot(x_vals, y_vals, label=label)
        for x_vals, y_vals, label in all_data2[name]:
            ax.plot(x_vals, y_vals, label=label, linestyle='--')
        ax.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

```
